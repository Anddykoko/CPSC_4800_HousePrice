#import libraries

```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Loading the dataset

```python
# Load the Titanic dataset
titanic_df = pd.read_csv("titanic.csv")
```

```python
# Display basic information about the dataset
titanic_info = titanic_df.info()

# Display the first few rows
titanic_head = titanic_df.head()

# Check for missing values
missing_values = titanic_df.isnull().sum()

# Get summary statistics
summary_statistics = titanic_df.describe()

# Check the distribution of survived passengers
survival_counts = titanic_df["Survived"].value_counts()

titanic_info, titanic_head, missing_values, summary_statistics, survival_counts
```

#Performing EDA

#setting the plotstyle

```python
# Set plot style
sns.set_style("whitegrid")
```

#Plot survival count

```python
# Plot survival count
plt.figure(figsize=(6, 4))
sns.countplot(x="Survived", data=titanic_df, color='skyblue', edgecolor='black')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

#Plot survival rate by class

```python
# Plot survival rate by class
plt.figure(figsize=(6, 4))
sns.countplot(x="Pclass", hue="Survived", data=titanic_df, palette="coolwarm")
plt.title("Survival Count by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

#Plot survival rate by sex

```python
# Plot survival rate by sex
plt.figure(figsize=(6, 4))
sns.countplot(x="Sex", hue="Survived", data=titanic_df, palette="coolwarm")
plt.title("Survival Count by Gender")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

#Age distribution by survival status

```python
# Age distribution by survival status
plt.figure(figsize=(8, 5))
sns.histplot(titanic_df, x="Age", hue="Survived", kde=True, bins=30, palette="coolwarm")
plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

#Fare distribution by survival status

```python
# Fare distribution by survival status
plt.figure(figsize=(8, 5))
sns.histplot(titanic_df, x="Fare", hue="Survived", kde=True, bins=30, palette="coolwarm")
plt.title("Fare Distribution by Survival Status")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.xlim(0, 300)  # Limiting for better visualization
plt.show()
```
